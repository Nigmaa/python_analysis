```python
country_mask = df1['國家'] == '日本'
df1[country_mask]
```

```python
#布林遮罩連續篩選搜尋資料用法
taiwan_mask = (df1['國家'] == '台灣') & (df1['新增確診數'] >= 100000)
df1[taiwan_mask]
```

```python
#也可以用query搜尋資料，達到與上面一樣效果
df1.query('國家=="台灣" and 新增確診數>=100000')
```

```python
#安裝matplotlib.pyplt模組
import matplotlib.pyplot as plt
#安裝matplotlib.font_manager模組中的fontManager實體並匯入
from matplotlib.font_manager import fontManager
import matplotlib as mlp
#使用fontManager的實體方法addfont，將ChineseFont.ttf匯入到matplotlib
fontManager.addfont('ChineseFont.ttf')
#使用matplotlib模組中的rc函數，將ChineseFont.ttf作為matplotlib的預設字型，此時就可以打中文了
mlp.rc('font', family='ChineseFont')
```

```python
figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1,1,1)
axes.set_xticks([1, 3, 5, 7, 9])
axes.set_yticks([-1, -0.5, 0, 0.5, 1])
plt.show()
```

```python
figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1,1,1)
axes.plot([11, 3, 5, 7, 9],marker='o',color='red', linestyle='')
plt.show()
```

```python
figure = plt.figure(figsize=(10,6))
ax1 = figure.add_subplot(2,1,1)
ax1.plot([11, 3, 5, 7, 9],marker='o',color='red', linestyle='')
ax2 = figure.add_subplot(2,1,2)
ax2.bar(['a', 'b', 'c', 'd', 'e'],[1, 2, 3, 4, 5])
plt.show()
```

```python
student1 = {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}
```

```python
subjects = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())
figure = plt.figure(figsize=(10,8))
axes = figure.add_subplot(1,1,1)
axes.set_title("小明與大雄成績表")
axes.set_xlabel("科目")
axes.set_ylabel("分數")
axes.plot(subjects,score1,'ro-',label="小明")
axes.plot(subjects,score2,'gx-.',label="大雄")
axes.grid(axis='y')
plt.show()
```
