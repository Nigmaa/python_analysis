```python
import pandas as pd
import numpy as np
# 建立一個包含 NaN 的 DataFrame
data = {'col1': [1, 2, np.nan, 4], 
        'col2': ['a', np.nan, 'c', 'd'], 
        'col3': [1.1, 2.2, 3.3, np.nan]}
df = pd.DataFrame(data)

print("原始 DataFrame:")
print(df)
```

```python
# 刪除任何包含 NaN 的列
df_cleaned = df.dropna()  # 預設是 axis=0, how='any'
print("\n刪除包含 NaN 的列後的 DataFrame：")
print(df_cleaned)
```

```python
# 直接在原 DataFrame 上刪除包含 NaN 的列
df.dropna(inplace=True)
print("\n在原 DataFrame 上刪除包含 NaN 的列後的 DataFrame：")
print(df)
```

```python
# 只有當該列的所有值都是 NaN 時，才刪除該列
data2 = {'col1': [1, 2, np.nan, 4], 
        'col2': ['a', np.nan, np.nan, 'd'], 
        'col3': [1.1, 2.2, np.nan, np.nan]}
df2 = pd.DataFrame(data2)
print("\n原始 DataFrame2：")
print(df2)
```

```python
df2_cleaned = df2.dropna(how='all')
print("\n刪除所有值都是 NaN 的列後的 DataFrame2：")
print(df2_cleaned)
```

```python
# 指定要檢查的列名列表
data3 = {'col1': [1, 2, np.nan, 4], 
        'col2': ['a', np.nan, 'c', 'd'], 
        'col3': [1.1, 2.2, 3.3, np.nan]}
df3 = pd.DataFrame(data3)
print("\n原始 DataFrame3:")
print(df3)
df3_cleaned = df3.dropna(subset=['col1','col3'])
print("\n刪除col1或col3包含NaN的列後的 DataFrame3：")
print(df3_cleaned)
```

```python
import pandas as pd

df = pd.read_csv('各鄉鎮市區人口密度.csv')
df.info()

#清理資料，將空資料刪除
df1 = df.dropna()
#檢查後面資料，發現有其他符號，例如:...
df1.tail()
#檢查前面資料，發現表首要刪除第二列(編號1的列)
df1.head()
#用info檢查type，發現數值資料顯示為字串，需轉換為數值
df1.info()
```

```python
#先將表首進行改名
df2 = df1.rename(columns=df1.iloc[0].to_dict())
df2
```

```python
#刪除多餘的表首
df3 = df2.drop(index=0)
df3.info()
```

```python
import numpy as np
#將數值資料轉換為數值，這邊有些資料有其他符號，例如:...之類的，如果直接轉換會出現錯誤，所以先定義一個函數，使用try_except函數檢查帶入的值是否可以轉換為數值，當不行跳出錯誤時，就不管帶入的值是什麼，都轉換為0
def str_to_int(value):
    try:
        value_int = int(value)
    except:
        value_int = 0
    return value_int
#土地面積是浮點數，所以要另外定義一個函數，與上面函數道理相同
def str_to_float(value):
   try:
        value_float = float(value)
   except:
        value_float = 0
   return value_float

df3['年底人口數'] = df3['年底人口數'].map(str_to_int)
df3['人口密度'] = df3['人口密度'].map(str_to_int)
df3['土地面積'] = df3['土地面積'].map(str_to_float)
```

```python
#將區域別資料剖析出縣市資料，因為南沙群島的區域別是4個字，所以我們可以用這個特性來剖析出縣市資料，如果區域別大於4個字(非南沙群島)，就取前3個字，如果小於等於4個字(南沙群島)，就取全部
def split_city(value):
    if value > '4':
        return value[:3]
    elif value <= '4':
        return value

#新增一個縣市欄位，並把剖析出來的縣市資料填入
df3['縣市'] = df3['區域別'].map(split_city)
df3
#重新選取我要的欄位資料
df4 = df3.reindex(columns=['縣市','年底人口數','人口密度','土地面積'])
df4
```

```python
#將數值資料進行區間化，這邊我們將人口密度分為高、中、低三個區間，使用cut函數，將人口密度分為高、中、低三個區間(裡面的bins參數能切3個區間)
df4['人口密度區間'] = pd.cut(df4['人口密度'], bins=3, labels=['低', '中', '高'])
df4
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#安裝matplotlib.pyplt模組
import matplotlib.pyplot as plt
#安裝matplotlib.font_manager模組中的fontManager實體並匯入
from matplotlib.font_manager import fontManager
import matplotlib as mlp
#使用fontManager的實體方法addfont，將ChineseFont.ttf匯入到matplotlib
fontManager.addfont('ChineseFont.ttf')
#使用matplotlib模組中的rc函數，將ChineseFont.ttf作為matplotlib的預設字型，此時就可以打中文了
mlp.rc('font', family='ChineseFont')
df_plot = df4[['縣市', '年底人口數']]

#按照縣市及年底人口數，畫出長條圖
figure = plt.figure(figsize=(12, 6))
picture = figure.add_subplot(1,1,1)
picture.set_xlabel('縣市')
picture.set_ylabel('年底人口數')
picture.set_title('各縣市人口數量分布圖')
picture.bar(df_plot['縣市'], df_plot['年底人口數'])
plt.xticks(rotation=45) # 旋轉 x 軸標籤 45 度
plt.show()


##按照縣市及年底人口數，畫出盒鬚圖
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_plot, x='縣市', y='年底人口數')
plt.xlabel('縣市')
plt.ylabel('年底人口數')
plt.title('各縣市人口數量盒鬚圖')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

scores = np.random.randint(50, 101, size=(50, 5))
df = pd.DataFrame(scores,
             columns=['國文', '英文', '數學', '地理', '歷史'],
             index=range(1,51)
             )
df.index.name = '學號'
df.columns.name = '科目'
df
sum_value = df.sum(axis=1)
mean_value = df.mean(axis=1)
df['總分'] = sum_value
df['平均'] = mean_value
#排序總分
df1 = df.sort_values(by='總分',ascending=False)
df1
#排名
rank_value = df1['總分'].rank(ascending=False,method='min')
df1['排名'] = rank_value
df1

#顯示每位學生的最高分
def max_score(row):
    less_60_count = len(row[row < 60])
    #新增4個新的欄位名稱及資料，分別為最低分、最高分、標準差、不及格的數量，這要用pandas的Series函數，並且用index參數指定新的欄位名稱
    return pd.Series([row.min(),row.max(),row.std(),less_60_count],index=['最低分','最高分','標準差','不及格的數量'])
#使用apply函數，將max_score函數帶入，並且axis=1，表示將每一列的資料都帶入
df2 = df1[['國文', '英文', '數學', '地理', '歷史']].apply(max_score,axis=1)
pd.concat([df1,df2],axis=1)
```
